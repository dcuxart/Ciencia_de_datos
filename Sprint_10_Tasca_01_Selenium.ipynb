```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.common.exceptions import NoSuchElementException

driver = webdriver.Chrome()
```

En primer lugar obtenemos la totalidad de las citas por cada una de las 3 páginas de las que conta el sitio web.

```python
url = ('http://quotes.toscrape.com/')
# driver.get(url)

quotes_selenium = []

try:
    while True:
        driver.get(url)
        
        quotes = driver.find_elements(by=By.XPATH, value="//div[@class='quote']/span[@class='text']")
        for quote in quotes:
            quotes_selenium.append(quote.text)

        next_button = driver.find_element(by=By.XPATH, value="//li[@class='next']/a")
        
        if not next_button.is_displayed():
            print("No more pages")
            break
            
        url = next_button.get_attribute('href')
        
except NoSuchElementException:
    print("No more pages")
```

```python
len(quotes_selenium)
```

Acto seguido hacemos exactamente lo mismo para los autores.

```python
url = 'http://quotes.toscrape.com/'

authors_selenium = []

try:
    while True:
        driver.get(url)
        
        authors = driver.find_elements(by=By.XPATH, value="//div[@class='quote']/span[2]/small")
        for author in authors:
            authors_selenium.append(author.text)

        next_buttons = driver.find_elements(by=By.XPATH, value="//li[@class='next']/a")

        if not next_buttons:
            print("No more pages")
            break

        next_button = next_buttons[0]
        if not next_button.is_displayed():
            print("No more pages")
            break
            
        url = next_button.get_attribute('href')
        
except NoSuchElementException:
    print("No more pages")
```

```python
len(authors_selenium)
```

Y finalmente para las etiquetas.

```python
url = 'http://quotes.toscrape.com/'

tags_selenium = []

try:
    while True:
        driver.get(url)
        
        tags = driver.find_elements(by=By.XPATH, value="//div[@class='quote']/div[@class='tags']")
        for tag in tags:
            tags_selenium.append(tag.text)

        next_buttons = driver.find_elements(by=By.XPATH, value="//li[@class='next']/a")

        if not next_buttons:
            print("No more pages")
            break

        next_button = next_buttons[0]
        if not next_button.is_displayed():
            print("No more pages")
            break
            
        url = next_button.get_attribute('href')
        
except NoSuchElementException:
    print("No more pages")
```

```python
len(tags_selenium)
```

Juntamos la información que hemos extraído en un mismo dataset.

```python
import pandas as pd

df_quotes_selenium = pd.DataFrame({'Author': authors_selenium, 'Quote': quotes_selenium, 'Tags': tags_selenium})
df_quotes_selenium
```

Pasamos ahora a realizar el mismo procedimiento pero para una tabla en un artículo de la wikipedia con los microestados en el mundo.

Primero obtenemos los años de la tabla.

```python
URL_wiki = "https://en.wikipedia.org/wiki/List_of_micronations"
```

```python
driver_2 = webdriver.Chrome()
driver_2.get(URL_wiki)

years_elements = driver_2.find_elements(by=By.XPATH, value="//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[2]")

years = []

for element in years_elements:
    years.append(element.text)

driver_2.quit()
len(years)
```

Acto seguido el nombre de los microestados.

```python
driver_2 = webdriver.Chrome()
driver_2.get(URL_wiki)

state_elements = driver_2.find_elements(by=By.XPATH, value="//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[3]")

states = []

for state in state_elements:
    states.append(state.text)

driver_2.quit()
len(states)
```

Y finalmente las descripciones.

```python
driver_2 = webdriver.Chrome()
driver_2.get(URL_wiki)

description_elements = driver_2.find_elements(by=By.XPATH, value="//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[4]")

descriptions = []

for description in description_elements:
    descriptions.append(description.text)

driver_2.quit()
len(descriptions)
```

Ahora que disponemos de toda la información de la tabla la juntamos en un mismo dataset.

```python
import pandas as pd

df_wiki_selenium = pd.DataFrame({'Year': years, 'State': states, 'Description': descriptions})
df_wiki_selenium
```
