#### Finalmente, algunas de estas variables están asignadas como PRIMARY KEY o FOREIGN KEY.

```python
#------- EXERCICI 2
SELECT person_name, person_country, person_dob
FROM tb_person
WHERE person_dod IS NULL
ORDER BY person_dob ASC
```

```python
#------- EXERCICI 3
SELECT genre_name, COUNT(movie_genre_id)
FROM tb_movie
JOIN tb_genre
	 ON tb_movie.movie_genre_id = tb_genre.genre_id
GROUP BY genre_name
ORDER BY COUNT(movie_genre_id) DESC;
```

```python
#------- EXERCICI 4
SELECT tb_person.person_name AS 'Nombre',
	COUNT(tb_movie_person.role_id) AS 'Roles distintos en un mismo film'
	FROM tb_movie_person
JOIN tb_person
	ON  tb_person.person_id = tb_movie_person.person_id
GROUP BY tb_person.person_name, tb_movie_person.movie_id
HAVING COUNT(tb_movie_person.role_id) > 1
ORDER BY COUNT(tb_movie_person.role_id) DESC
```
